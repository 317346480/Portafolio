El archivo **water_potability.csv** contiene métricas de calidad del agua para **3276 cuerpos de agua diferentes**, los cuales se describen a partir de **14 atributos/columnas**, los cuales son:


1. **pH**: pH del agua (de 0 a 14).
2. **Dureza**: Capacidad del agua para precipitar jabón en mg/L.
3. **Sólidos**: Sólidos totales disueltos en ppm.
4. **Cloraminas**: Cantidad de cloraminas en ppm.
5. **Sulfato**: Cantidad de sulfatos disueltos en mg/L.
6. **Conductividad**: Conductividad eléctrica del agua en μS/cm.
7. **Carbono_orgánico**: Cantidad de carbono orgánico en ppm.
8. **Trihalometanos**: Cantidad de trihalometanos en μg/L.
9. **Turbidez**: Medida de la propiedad de emisión de luz del agua en NTU.
10. **Potabilidad**: Indica si el agua es segura para el consumo humano. Potable -1 y No potable -0

Donde las *unidades de medida* corresponden a:
- **ppm**: partes por millón
- **μg/L**: microgramos por litro
- **mg/L**: miligramos por litro

Una observación importante es que todos los datos de este dataset son de tipo numérico, esto permite tomar la decisión de que los valores nulos o NaN puedan ser sustituidos por el promedio de los valores de la categoría correspondiente, pero tomando ciertas consideraciones que se abordan en la sección *Limpieza del dataset*.

##Limpieza del dataset

Se eliminarán los registros de los que se carezca toda la informarción, es decir, que todos los valores de sus atributos sean NaN, de igual forma, se eliminarán los atributos de lo que no se tenga ningún dato para ningun registro.

```python
df_dropped=df.dropna(axis=0, how='all') #elimina filas donde todos los valores son NaN
df_dropped=df.dropna(axis=1, how='all') #elimina columnas donde todos los valores son NaN
df_dropped
```

```python
nulos = df.isna().sum().sum()
print("Existen en total",nulos,"NaNs")
```

Una vez hecho lo anterior, se sustituirán el restante de los valores NaN con el promedio de los valores de cada columna correspondiente, dado que las columnas cuyos valores además de ser numéricos, son categóricos de forma binaria o dentro de un rango (ph, turbidez y potabilidad) no tienen ningún valor tipo NaN, por lo que esta acción de sustitución no representa un daño a la calidad de la información.

```python
promedios = df.mean()
promedios
#Colocaremos el valor promedio en los NaNs
df= df.fillna(promedios)
```

```python
nulos = df.isna().sum().sum()
print("Existen en total",nulos,"NaNs")
```

Después de aplicar estos procesos de limpieza, con la ejecución de la celda anterior que hace un recuento de los NaNs en el df, podemos afirmar que nuestro df está libre de NaNs y podemos trabajar óptimamente con él
